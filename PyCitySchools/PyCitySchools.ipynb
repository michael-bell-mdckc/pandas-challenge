```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(school_data, student_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# total schools
total_schools = school_data["school_name"].count()

# total students
total_students_size = school_data["size"].sum()

# total budget
total_schools_budget = school_data["budget"].sum()

# average math score
average_math_score = student_data["math_score"].mean()

# average reading score
average_reading_score = student_data["reading_score"].mean()

# overall passing rate
passing_rate = (average_math_score + average_reading_score) / 2

# passing math score
passing_math = student_data.loc[student_data["math_score"] >= 70]
passing_math_students = passing_math["student_name"].count()
passing_math_students_percentage = (passing_math_students / total_students_size) *100

# passing reading score
passing_reading = student_data.loc[student_data["reading_score"] >= 70]
passing_reading_students = passing_reading["student_name"].count()
passing_reading_students_percentage = (passing_reading_students / total_students_size) *100

# summary dataframe with formatting
summary_table_01 = pd.DataFrame({"Total Schools": [total_schools],
                                 "Total Students": [total_students_size],
                                 "Total Budget": [total_schools_budget],
                                 "Average Math Score": [average_math_score],                              
                                 "Average Reading Score": [average_reading_score], 
                                 "% Passing Math": [passing_math_students_percentage],
                                 "% Passing Reading": [passing_reading_students_percentage],
                                 "% Overall Passing Rate": [passing_rate]})

summary_table_01["Total Students"] = summary_table_01["Total Students"].astype(float).map("{:,.0f}".format)
summary_table_01["Total Budget"] = summary_table_01["Total Budget"].astype(float).map("${:,.2f}".format)

summary_table_01
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# each school name
grouped_school_data_complete = school_data_complete.groupby(["school_name"])

# each school type
each_school_type = grouped_school_data_complete["type"].unique()

# each school total students
each_school_students = grouped_school_data_complete["size"].mean()

# each school total budget
each_school_budget = grouped_school_data_complete["budget"].mean()

# each school per student budget
each_school_student_budget = each_school_budget / each_school_students

# each school average math score
each_school_average_math_score = grouped_school_data_complete["math_score"].mean()

# each school average reading score
each_school_average_reading_score = grouped_school_data_complete["reading_score"].mean()

# each school passing math score
passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
each_school_passing_math = passing_math.groupby(["school_name"])
each_school_passing_math_students = each_school_passing_math["student_name"].count()
each_school_passing_math_students_percentage = (each_school_passing_math_students / each_school_students) *100

# each school passing reading score
passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
each_school_passing_reading = passing_reading.groupby(["school_name"])
each_school_passing_reading_students = each_school_passing_reading["student_name"].count()
each_school_passing_reading_students_percentage = (each_school_passing_reading_students / each_school_students) *100

# each school overall passing rate
each_school_passing_rate = (each_school_passing_math_students_percentage + each_school_passing_reading_students_percentage) / 2

# summary dataframe with formatting
summary_table_02 = pd.DataFrame({"School Type": each_school_type,
                                 "Total Students": each_school_students,
                                 "Total School Budget": each_school_budget,
                                 "Per Student Budget": each_school_student_budget,
                                 "Average Math Score": each_school_average_math_score,                              
                                 "Average Reading Score": each_school_average_reading_score, 
                                 "% Passing Math": each_school_passing_math_students_percentage,
                                 "% Passing Reading": each_school_passing_reading_students_percentage,
                                 "% Overall Passing Rate": each_school_passing_rate})

summary_table_02["Total Students"] = summary_table_02["Total Students"].astype(float).map("{:,.0f}".format)
summary_table_02["Total School Budget"] = summary_table_02["Total School Budget"].astype(float).map("${:,.2f}".format)
summary_table_02["Per Student Budget"] = summary_table_02["Per Student Budget"].astype(float).map("${:,.2f}".format)

# sort top schools
top_schools = summary_table_02.sort_values(["% Overall Passing Rate"], ascending=False)
top_schools.index.name = None
top_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
# sort bottom schools
bottom_schools = summary_table_02.sort_values(["% Overall Passing Rate"], ascending=True)
bottom_schools.index.name = None
bottom_schools.head()
```

```python
# Bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

summary_table_05 = pd.DataFrame({"Per Student Budget": each_school_student_budget,
                                 "Average Math Score": each_school_average_math_score,
                                 "Average Reading Score": each_school_average_reading_score,
                                 "% Passing Math": each_school_passing_math_students_percentage,
                                 "% Passing Reading": each_school_passing_reading_students_percentage,
                                 "% Overall Passing Rate": each_school_passing_rate})

# slice the data and place it into bins
summary_table_05["Spending Ranges (Per Student)"] = pd.cut(summary_table_05["Per Student Budget"], spending_bins, labels=group_names)

# create the groupby object
summary_table_05_grouped = summary_table_05.groupby("Spending Ranges (Per Student)").mean()

# each column within the groupby object
summary_table_05_grouped_output = summary_table_05_grouped[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]

# summary dataframe
summary_table_05_grouped_output
```
